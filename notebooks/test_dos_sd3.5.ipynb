```python
import torch
import argparse
from diffusers import StableDiffusion3Pipeline

import configs.envs
from run_benchmark import default, dos_sd3_5
from utils.visualizer import concatenate_images_with_spacing
```

```python
device = "cuda:0"
pipe = StableDiffusion3Pipeline.from_pretrained(
    "stabilityai/stable-diffusion-3.5-large", torch_dtype=torch.bfloat16)
pipe.enable_model_cpu_offload(device=device)
```

```python
seed = 5
prompt = "a carrot and an ice cream cone"
objects = ["carrot", "ice cream cone"]
image_size = 256
padding = 10

args = argparse.Namespace(
    device=device,
    lambda_sep=1.0
)
image_default = default(pipe=pipe, seed=seed, target_prompt=prompt, target_objects=objects, args=args).resize((image_size,image_size))
image_embedding_calibration = dos_sd3_5(pipe=pipe, seed=seed, target_prompt=prompt, target_objects=objects,args=args).resize((image_size,image_size))
concatenated_image = concatenate_images_with_spacing(images=[image_default, image_embedding_calibration], spacing=padding)
display(concatenated_image)
```
