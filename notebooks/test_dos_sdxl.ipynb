```python
import torch
import argparse
from diffusers import StableDiffusionXLPipeline

import configs.envs
from run_benchmark import default, dos_sdxl
from utils.visualizer import concatenate_images_with_spacing
```

```python
device = "cuda:0"
pipe = StableDiffusionXLPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, variant="fp16", use_safetensors=True
).to(device)
```

```python
seed = 32
prompt = "a mushroom and an umbrella"
objects = ["mushroom", "umbrella"]
image_size = 256
padding = 10

args = argparse.Namespace(
    device=device,
    lambda_sep=1.0
)
image_default = default(pipe=pipe, seed=seed, target_prompt=prompt, target_objects=objects, args=args).resize((image_size,image_size))
image_embedding_calibration = dos_sdxl(pipe=pipe, seed=seed, target_prompt=prompt, target_objects=objects,args=args).resize((image_size,image_size))
concatenated_image = concatenate_images_with_spacing(images=[image_default, image_embedding_calibration], spacing=padding)
display(concatenated_image)
```
